```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
z=tf.constant([-128.0, 0.0, 128.0], dtype=tf.float32)
```

```python
 print('Input type:', z)
```

```python
tf.math.sigmoid(z)
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.linspace(-5, 5, 50)
print('Input type:', a)
#z_hat= tf.nn.sigmoid(a, name ='sigmoid')
z_hat=tf.math.sigmoid(a)
```

```python
print('Output type:', z_hat)
```

```python
plt.plot(z_hat,color = 'red', marker = "o") 
plt.title("tensorflow sigmoid") 
plt.xlabel("X") 
plt.ylabel("Y") 
  
plt.show()
```

```python
z_hat=tf.math.tanh(a)
```

```python
plt.plot(z_hat,color = 'blue', marker = "o") 
plt.title("Tensorflow Tanh") 
plt.xlabel("X") 
plt.ylabel("Y") 
  
plt.show()
```

```python
z_hat=tf.nn.relu(a)
```

```python
plt.plot(z_hat,color = 'pink', marker = "o") 
plt.title("Tensorflow ReLu") 
plt.xlabel("X") 
plt.ylabel("Y") 
  
plt.show()
```
