```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
```

```python
data = np.random.random((1000,100))
labels = np.random.randint(2,size=(1000,1)) # Predict this one
model = Sequential()
```

```python
#labels
```

```python
model.add(Dense(32,
 activation='relu', 
 input_dim=100))
model.add(Dense(1, activation='sigmoid'))
```

```python
'''
model.compile(optimizer='sgd',
 loss='mean_squared_error',  #   binary_crossentropy can be used for classifications
 metrics=['accuracy'])

'''

# More about losses at https://keras.io/api/losses/regression_losses/#meansquarederror-class 

model.compile(
    optimizer='sgd',
    loss='mse',
    metrics=[tf.keras.metrics.MeanSquaredError()])
```

```python
model.fit(data,labels,epochs=10,batch_size=32)
```

```python
model.fit(data,labels,epochs=200,batch_size=32)
```

```python
# epoch is one forward and backward pass and in this case the loss reduces when epochs changed from 10 to 200
```

```python
# Calling `save('my_model.h5')` creates a h5 file `my_model.h5`.
model.save("AnirbanFakeData_h5_model.h5")

```

```python
import keras
# It can be used to reconstruct the model identically.
reconstructed_model = keras.models.load_model("AnirbanFakeData_h5_model.h5")

# https://www.tensorflow.org/guide/keras/save_and_serialize 
```

```python
test_input=data
test_target=labels
# Let's check:
np.testing.assert_allclose(
    model.predict(test_input), reconstructed_model.predict(test_input)
)

# The reconstructed model is already compiled and has retained the optimizer
# state, so training can resume:
reconstructed_model.fit(test_input, test_target)
```
